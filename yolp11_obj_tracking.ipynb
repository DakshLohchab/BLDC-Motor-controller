**Object Detection**

```python
from ultralytics import YOLO

model = YOLO('yolo11m.pt')

result = model('/content/6387-191695740_small.mp4', save=True, save_txt=False, save_conf=False)
```

```python
result = model('/content/6387-191695740_small.mp4',classes=0,save=True, save_txt=False, save_conf=False)
```

**TRACKINGa**

```python
# Bot Sort

results = model.track(source='/content/6387-191695740_small.mp4',persist=True,save=True)
```

```python
#Byte Track

result = model.track('/content/6387-191695740_small.mp4',persist=True,save=True,classes=0,show=True,tracker='bytetrack.yaml')
```
